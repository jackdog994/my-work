### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
np.round(users.groupby('occupation').mean()['age'])
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users_gender_occ = users.groupby(['gender','occupation']).count()
gender_ratio = users_gender_occ.loc['M']['user_id']/(users_gender_occ.loc['M']['user_id'] + users_gender_occ.loc['F']['user_id']) 
gender_ratio.fillna(value=1.00,inplace=True)
gender_ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].aggregate(['max', 'min'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
np.round(users.groupby(['gender','occupation'])['age'].mean())
```

### Step 8.  For each occupation present the percentage of women and men

```python
ratio_men = users_gender_occ.loc['M']['user_id']/(users_gender_occ.loc['M']['user_id'] + users_gender_occ.loc['F']['user_id'])
ratio_men.rename('M %',inplace=True).fillna(value=1.00,inplace=True)
ratio_women = users_gender_occ.loc['F']['user_id']/(users_gender_occ.loc['M']['user_id'] + users_gender_occ.loc['F']['user_id'])
ratio_women.rename('F %',inplace=True).fillna(value=0.00,inplace=True)
ratio = pd.concat([ratio_men,ratio_women],axis=1)
ratio
```
