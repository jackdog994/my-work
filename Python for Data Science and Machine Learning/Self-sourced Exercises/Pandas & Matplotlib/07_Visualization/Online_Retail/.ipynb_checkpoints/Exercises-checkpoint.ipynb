### Introduction:



### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Visualization/Online_Retail/Online_Retail.csv). 

### Step 3. Assign it to a variable called online_rt

```python
online_rt = pd.read_csv('Online_Retail.csv')
```

### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
top10 = online_rt.groupby(['Country']).sum()
```

```python
top10 = top10[top10.index!='United Kingdom'].sort_values(by='Quantity',ascending=False).drop(['UnitPrice','CustomerID'],axis=1).head(10)
```

```python
top10.plot(kind='bar')
plt.ylabel('Quantity')
```

### Step 5.  Exclude negative Quatity entries

```python
online_rt = online_rt[online_rt['Quantity']>0]
```

### Step 6. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries

```python
country_cust = online_rt.groupby(['Country','CustomerID']).sum()
country_cust_scatter, axes = plt.subplots(1, 3, sharey=True,figsize=(13,4))
axes[0].scatter(country_cust.loc['Netherlands']['Quantity'],country_cust.loc['Netherlands']['UnitPrice'])
axes[1].scatter(country_cust.loc['EIRE']['Quantity'],country_cust.loc['EIRE']['UnitPrice'])
axes[2].scatter(country_cust.loc['Germany']['Quantity'],country_cust.loc['Germany']['UnitPrice'])
axes[0].set_ylabel('Unit Price')
axes[1].set_xlabel('Quantity')
axes[0].set_title('Netherlands')
axes[1].set_title('EIRE')
axes[2].set_title('Germany')
country_cust_scatter.suptitle('Quantity per Unit Price for Top 3 Countries')
```
